```python
import pyttsx3
import speech_recognition as sr
import webbrowser


def speaktext(text):
    friend=pyttsx3.init()
    friend.say(text)
    friend.runAndWait()

speaktext('Hi i am jupyter')


def getaudio():
    r = sr.Recognizer()
    with sr.Microphone() as source:
        print('listening...')
        audio = r.listen(source)
        MyText = r.recognize_google(audio)
        return MyText
```

```python
import random
joke=['what did the fish say when he swam into a wall, Dam','Did you hear about the fire in the shoe factory, ten thousand soles were lost the police said some heels started it','What do you call a fake noodle, an impasta','Did you hear the story about the claustrophobic astronaut, he just needed some space']
jokechoice=random.choice(joke)
def talk_to_jupyter():
    while True:
        query=getaudio()
        if 'hello' in query:
            speaktext('hii how are you')
        elif 'hi how are you' in query:
            speaktext('i am fine How about you? ')
        elif 'bye' in query:
            speaktext('Bye See you Again')
            break
        elif 'name' in query:
            speaktext('My name is Jupyter what is your name')
        elif 'help' in query:
            speaktext('WHat can i help you with')
        elif 'joke' in query:
            speaktext('Here is a joke')
            speaktext(jokechoice)
            break
        else:
            exit()
def voiceassistant():
    while True:
        query=getaudio()
        if 'open Google' in query:
            speaktext('What do you want to search for?')
            search_query = getaudio()
            speaktext(f'searching for {search_query} on Google')
            webbrowser.open(f'https://www.google.co.in/search?q={search_query}')
        elif 'open YouTube' in  query:
            speaktext('What do you want to search for?')
            search_query1 = getaudio()
            speaktext(f'searching for {search_query1} on YouTube')
            webbrowser.open(f'https://www.youtube.com/results?search_query={search_query1}')
        elif 'open calculator' in query:
            speaktext('opening calculator')
            webbrowser.open('https://www.calculator.net/')
        elif 'open scratch' in query:
            speaktext('opening scratch')
            webbrowser.open('https://scratch.mit.edu/projects/editor/?tutorial=getStarted')
        elif 'open okru' in query:
            speaktext('opening ok ru')
            webbrowser.open('https://ok.ru/video//c5959553')
        elif 'open Paper IO' in query:
            speaktext('opening paper io')
            webbrowser.open('https://paper-io.com')
        elif 'open maps' in query:
            speaktext('opening google maps')
            webbrowser.open('https://www.google.com/maps/@53.800651,-4.064941,6z')
            break
        else:
            exit()
```

```python
# tkinter GUI
from tkinter import *
root= Tk()
#Make a Canvas (i.e, a screen for your project
canvas1 = tk.Canvas(root, width = 500, height = 300,bg='deeppink')
canvas1.pack()

# load the image file
img = PhotoImage(file='mic1.png')

# create the image on the canvas
canvas1.create_image(250, 150, image=img)

b1 = Button(root,text = "MIC",command = voiceassistant,activeforeground = "blue",activebackground = "pink",pady=10)
canvas1.create_window(250, 230, window=b1)
b2 = Button(root,text = "TALK WITH JUPYTER",command = talk_to_jupyter, activeforeground = "blue",activebackground = "pink",pady=10)
canvas1.create_window(250, 65, window=b2)
root.mainloop()
```
